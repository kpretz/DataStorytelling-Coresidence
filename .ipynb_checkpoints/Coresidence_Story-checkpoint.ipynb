```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import geopandas as gpd
import geodatasets
```

```python
data = pd.read_csv('A0_CORESIDENCE_NATIONAL_DATASET.csv')
codebook = pd.read_csv('CODEBOOK.csv')

name_dict = dict(zip(codebook['Code'], codebook['Description']))

data.rename(name_dict, axis=1, inplace=True)
data.rename({'Country ISO nomenclature 3 digits': 'ISO_A3'}, axis=1, inplace=True)
data.head()
```

### How many people are in each house by continent? How does it change over time? 

- Africa: Fairly flat mix of household size
- Europe: More 2-person households at the expense of 3- and 4-person households. Less kids?
- Latin America: Less 7+ person households - possible fewer multi-family or multi-generational housing situations?
- North America: More 1- and 2-person households at the expense of larger households. Less kids? More single people?
- Asia: fairly flat
- Oceania: biggest shift due to Australia being added to survey (above)

```python
proportion_col_names = ['Proportion of 1-person households','Proportion of 2-persons households',
 'Proportion of 3-persons households','Proportion of 4-persons households', 'Proportion of 5-persons households',
 'Proportion of 6-persons households', 'Proportion of 7-persons households', 'Proportion of 8-persons households',
 'Proportion of 9-persons households', 'Proportion of 10-persons households', 'Proportion of 11+persons households'] 
data_proportions_year = data[proportion_col_names + ['Census round', 'Continent']]
data_proportions_year = data_proportions_year.groupby(['Census round', 'Continent']).mean().reset_index()
# data_proportions_year

for cont in data_proportions_year['Continent'].unique():

    data_proportions_year_sub_cont = data_proportions_year.loc[data_proportions_year['Continent'] == cont, :]
    col = sns.color_palette("Set3", 11)
    f, ax = plt.subplots(1, 1)
    ax.stackplot(data_proportions_year_sub_cont['Census round'], data_proportions_year_sub_cont[proportion_col_names].T, colors=col)
    ax.set_title(cont)
    ax.legend(proportion_col_names, bbox_to_anchor=(1.05, 1.))
    
# plt.legend(data_proportions_year.columns)
```

### What sorts of correlations are there? 

- Average household size:
    - more children --> bigger household size: obvious
    - more seniors (65+) --> smaller household size: interesting; suggests seniors are often living alone or with one other person. Curious how senior centers are handled in this data
    - higher GDI --> smaller household size: obvious
    - more years of schooling --> smaller household size: obvious
- population positively correlated with all variables
- male heads of house
    - more 1-person households --> fewer male heads of house: more women living alone than men?
    - bigger household size --> more male heads: if there is a male head, more children?
    - number of seniors in house --> fewer male heads: women live longer, but also senior women are the heads more often?
    - higher general income measure (GDI, mean schooling, life expectancy) --> more male heads of house

```python
corr_cols = ['Proportion of 1-person households', 'Proportion of 5-persons households', 'Proportion of 10-persons households',
            'Average household size', 'Average number of 0-4 children in the household', 'Average number of persons aged 65+ in the household',
             'Proportion of male-headed households','Population','Fertily rate', 'Life expectancy', 'Gross Domestic income', 'Mean years of schooling']
sns.heatmap(data[corr_cols].corr(numeric_only=True), vmin=-1, vmax=1, annot=True)
```
